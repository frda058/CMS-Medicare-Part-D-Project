# Import .save file

```python
# Random Forest
file_RF = open('RF_grid.save', 'rb')
X_test_RF, y_test_RF, model_RF, f05_RF, confusionmatrix_RF, accuracy_RF, ftr_names_RF = pickle.load(file_RF)
file_RF.close()


# Mean and Standard Deviation of F05 and Accuracy Score

mean_f05_RF = np.mean(f05_RF)
std_f05_RF = np.std(f05_RF)
mean_accuracy_RF = np.mean(accuracy_RF)
std_accuracy_RF = np.std(accuracy_RF)
```

```python
# XGB Classifier
file_XGB = open('XGB_grid.save', 'rb')
X_test_XGB, y_test_XGB, model_XGB, f05_XGB, confusionmatrix_XGB, accuracy_XGB, ftr_names_XGB = pickle.load(file_XGB)
file_XGB.close()


# Mean and Standard Deviation of F05 and Accuracy Score

mean_f05_XGB = np.mean(f05_XGB)
std_f05_XGB = np.std(f05_XGB)

mean_accuracy_XGB = np.mean(accuracy_XGB)
std_accuracy_XGB = np.std(accuracy_XGB)

mean_accuracy_XGB
```

```python
# SVC 
file_SVC = open('SVC_grid.save', 'rb')
X_test_SVC, y_test_SVC, model_SVC, f05_SVC, confusionmatrix_SVC, accuracy_SVC, ftr_names_SVC = pickle.load(file_SVC)
file_SVC.close()


# Mean and Standard Deviation of F05 and Accuracy Score

mean_f05_SVC = np.mean(f05_SVC)
std_f05_SVC = np.std(f05_SVC)
mean_accuracy_SVC = np.mean(accuracy_SVC)
std_accuracy_SVC = np.std(accuracy_SVC)
```

```python
# LogReg w/ L1 Penalty
file_LRP = open('LogReg_L1_grid.save', 'rb')
X_test_LRP, y_test_LRP, model_LRP,confusionmatrix_LRP, f05_LRP,  accuracy_LRP, ftr_names_LRP = pickle.load(file_LRP)
file_LRP.close()


# Mean and Standard Deviation of F05 and Accuracy Score

mean_f05_LRP = np.mean(f05_LRP)
std_f05_LRP = np.std(f05_LRP)

mean_accuracy_LRP = np.mean(accuracy_LRP)
std_accuracy_LRP = np.std(accuracy_LRP)
```

```python
# LogReg w/o Penalty
file_LR = open('LogReg_noPenalty_grid.save', 'rb')
X_test_LR, y_test_LR, model_LR,confusionmatrix_LR, f05_LR,  accuracy_LR, ftr_names_LR = pickle.load(file_LR)
file_LR.close()


# Mean and Standard Deviation of F05 score

mean_f05_LR = np.mean(f05_LR)
std_f05_LR = np.std(f05_LR)
mean_accuracy_LR = np.mean(accuracy_LR)
std_accuracy_LR = np.std(accuracy_LR)
```

# Confusion Matrix

```python
confusionmatrix_RF_avg = np.sum([confusionmatrix_RF[0] , 
                               confusionmatrix_RF[1], 
                               confusionmatrix_RF[2],
                               confusionmatrix_RF[3],
                               confusionmatrix_RF[4]], axis = 0)/5

confusionmatrix_RF_avg
```

```python
confusionmatrix_SVC_avg = np.sum([confusionmatrix_SVC[0] , 
                               confusionmatrix_SVC[1], 
                               confusionmatrix_SVC[2],
                               confusionmatrix_SVC[3],
                               confusionmatrix_SVC[4]], axis = 0)/5

confusionmatrix_SVC_avg
```

```python
confusionmatrix_XGB_avg = np.sum([confusionmatrix_XGB[0] , 
                               confusionmatrix_XGB[1], 
                               confusionmatrix_XGB[2],
                               confusionmatrix_XGB[3],
                               confusionmatrix_XGB[4]], axis = 0)/5

confusionmatrix_XGB_avg
```

```python
confusionmatrix_LRP

confusionmatrix_LRP_avg = np.sum([confusionmatrix_LRP[0] , 
                               confusionmatrix_LRP[1], 
                               confusionmatrix_LRP[2],
                               confusionmatrix_LRP[3],
                               confusionmatrix_LRP[4]], axis = 0)/5

confusionmatrix_LRP_avg
```

```python
confusionmatrix_LR

confusionmatrix_LR_avg = np.sum([confusionmatrix_LR[0] , 
                               confusionmatrix_LR[1], 
                               confusionmatrix_LR[2],
                               confusionmatrix_LR[3],
                               confusionmatrix_LR[4]], axis = 0)/5

confusionmatrix_LR_avg
```

```python
# Normalize Confusion Matrix
cm1_norm = normalize(confusionmatrix_XGB_avg, axis=1, norm='l1')
cm2_norm = normalize(confusionmatrix_RF_avg, axis=1, norm='l1')
cm3_norm = normalize(confusionmatrix_SVC_avg, axis=1, norm='l1')
cm4_norm = normalize(confusionmatrix_LR_avg, axis=1, norm='l1')
cm5_norm = normalize(confusionmatrix_LRP_avg, axis=1, norm='l1')
```

```python
from sklearn.preprocessing import normalize
import seaborn as sns
os.chdir('../figures')
# Labels:
#'Endocrinology': 0, 'Nephrology': 1, 'Ophthalmology'
ticks = ['Endocrinology', 'Nephrology', 'Ophthalmology']
sns.set(font_scale=2)


plt.figure(figsize=(15,20))


plt.subplot(5,1,1)
g1 = sns.heatmap(cm1_norm,cmap="YlGnBu",cbar=True, annot=True,fmt='.2f', annot_kws={'size': 25},
                 yticklabels=ticks, xticklabels=ticks,vmin=0, vmax=1)
g1.set_xlabel('Predicted',fontsize=25)
g1.set_ylabel('True',fontsize=25)
g1.set_xticklabels(['Endocrinology', 'Nephrology', 'Ophthalmology'], rotation=0)
g1.set_yticklabels(['Endocrinology', 'Nephrology', 'Ophthalmology'], rotation=0)
g1.set_title('XGBoost', fontweight='bold',fontsize=25)

plt.subplot(5,1,2)
g2 = sns.heatmap(cm2_norm,cmap="YlGnBu",cbar=True, annot=True,fmt='.2f', annot_kws={'size': 25},
                yticklabels=ticks, xticklabels=ticks,vmin=0, vmax=1)
g2.set_xlabel('Predicted',fontsize=25)
g2.set_ylabel('True',fontsize=25)
g2.set_xticklabels(['Endocrinology', 'Nephrology', 'Ophthalmology'], rotation=0)
g2.set_yticklabels(['Endocrinology', 'Nephrology', 'Ophthalmology'], rotation=0)
g2.set_title('Random Forest', fontweight='bold',fontsize=25)

plt.subplot(5,1,3)
g3 = sns.heatmap(cm3_norm,cmap="YlGnBu",cbar=True, annot=True,fmt='.2f', annot_kws={'size': 25},
                 yticklabels=ticks, xticklabels=ticks, vmin=0, vmax=1)
g3.set_xlabel('Predicted',fontsize=25)
g3.set_ylabel('True',fontsize=25)
g3.set_xticklabels(['Endocrinology', 'Nephrology', 'Ophthalmology'], rotation=0)
g3.set_yticklabels(['Endocrinology', 'Nephrology', 'Ophthalmology'], rotation=0)
g3.set_title('Support Vector Classifier', fontweight='bold',fontsize=25)

plt.subplot(5,1,4)
g4 = sns.heatmap(cm4_norm,cmap="YlGnBu",cbar=True, annot=True,fmt='.2f', annot_kws={'size': 25},
                 yticklabels=ticks, xticklabels=ticks,vmin=0, vmax=1)
g4.set_xticklabels(['Endocrinology', 'Nephrology', 'Ophthalmology'], rotation=0)
g4.set_yticklabels(['Endocrinology', 'Nephrology', 'Ophthalmology'], rotation=0)
g4.set_xlabel('Predicted',fontsize=25)
g4.set_ylabel('True',fontsize=25)
g4.set_title('Logistic Regression', fontweight='bold',fontsize=25)

plt.subplot(5,1,5)
g5 = sns.heatmap(cm5_norm,cmap="YlGnBu",cbar=True, annot=True,fmt='.2f', annot_kws={'size': 25},
                 yticklabels=ticks, xticklabels=ticks,vmin=0, vmax=1)
g5.set_xticklabels(['Endocrinology', 'Nephrology', 'Ophthalmology'], rotation=0)
g5.set_yticklabels(['Endocrinology', 'Nephrology', 'Ophthalmology'], rotation=0)
g5.set_xlabel('Predicted',fontsize=25)
g5.set_ylabel('True',fontsize=25)
g5.set_title('Logistic Regression (L1)', fontweight='bold',fontsize=25)


#plt.subplots_adjust(top=0.30)
#plt.subplots_adjust(hspace=0.7)
plt.tight_layout()
#plt.suptitle('Normalized Confusion Matrices for 5 ML Models',fontsize=20, fontweight='bold')
plt.savefig('Normalized Confusion Matrix (Avg) for All Models.png', dpi=1200)

plt.show()
```

# Accuracy

```python
# combined means and std dev

mean_score = [mean_accuracy_XGB, mean_accuracy_RF, mean_accuracy_SVC,mean_accuracy_LRP, mean_accuracy_LR]
std_score = [std_accuracy_XGB, std_accuracy_RF,  std_accuracy_LRP, std_accuracy_LRP, std_accuracy_LR]
models = ['XGBoost', 'Random Forest','SVC', 'Log. Reg (L1)', 'Log. Reg',]

# Plot Bar chart

#plt.bar(models, mean_score)

# Plot error bar
t = 'Baseline Accuracy = 0.34'
sns.set(font_scale=1.2)
plt.errorbar(models, mean_score, yerr = std_score,fmt='o',ecolor = 'red',color='black',capsize = 4,)
plt.axhline(y = 0.34, color = 'r', linestyle = '-')
plt.ylim(0, 1)
plt.text(1, 0.35, t, ha='left', rotation=0, wrap=True)

#plt.bar(models, mean_score, std_score, linestyle='None', marker='.', capsize = 4, markerfacecolor='g')
plt.title('Accuracy Score for Different ML Algorithms', fontweight = 'bold')
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.savefig("Accuracy Score for 4 ML Models.png", dpi=1200)
plt.show()
```
